```python
import json

with open('cached_data/acs5_2020_fields.json') as f:
    fields = json.load(f)
    
# with open('cached_data/acs5_2020_tables.json') as f:
#     tables = json.load(f)
    
feat_codes = []

for k in fields.keys():
    if k=='ucgid' or k[0]!='B' or k[-1]!='E':
        continue
    else:
        feat_codes.append(k)

print(len(feat_codes), feat_codes[:10])
```

```python
from tqdm.notebook import trange, tqdm
```

```python
%%time

import pandas as pd

# feat_names = feat_codes[:2]
# feat_name = feat_codes[:10]
# feat_names = feat_codes[:10]
# feat_names = feat_codes[:10]
feat_names = feat_codes

fips = '183' ## Wake county

with open('wake_county_feats.txt', 'w') as f:
    
    for feat in tqdm(feat_names, total=len(feat_names)):
        try:
            res = c.acs5.get(feat, geo={'for': f'county:183', 'in': 'state:37'})
            line = [feat,str(int(res[0][feat]))]
            line = (',').join(line)
            f.write(f"{line}\n")
        except Exception as e:
            print(e)
            continue
        f.flush()
```
