```python
df['County'] = df.County.apply(lambda x: x[:x.index(' ')])  

dff = df[df.Date.isin([20220205, 20220212, 20220219])]
dff = dff[(dff['Pop'] < 3600.5) 
         | (dff['Pop'] > 1000000) | (dff['County'] == 'Orange')
        | (dff['County'] == 'Wake')]

dff
```

```python
dff['Proj_state_inc_deaths'] = dff.Proj_state_inc_deaths.apply(lambda x: int(x))
dff['True_county_inc_deaths_tm_4'] = dff.True_county_inc_deaths_tm_4.apply(lambda x: int(x))
dff['True_county_inc_deaths_tm_5'] = dff.True_county_inc_deaths_tm_5.apply(lambda x: int(x))
dff['True_county_inc_deaths_tm_6'] = dff.True_county_inc_deaths_tm_6.apply(lambda x: int(x))
dff['True_county_inc_deaths'] = dff.True_county_inc_deaths.apply(lambda x: int(x))

dff
```
