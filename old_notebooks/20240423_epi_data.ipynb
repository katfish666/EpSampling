```python
import os

dfs = []
tot = 0

for f in os.listdir(wdir):
    if 'COVID' in f: continue
    print(f)
    fname = wdir+f
    df = pd.read_csv(fname)
    tot += len(df.columns)
    display(df)
    
    dfs.append(df)
    
print(tot)
```

```python
first_df = pd.read_csv(wdir+'NC_county_acs5_2021_age_sex.csv')
```

```python
for each in dfs:
    df = pd.merge(first_df, each, on='GEOID')
    first_df = df
```
