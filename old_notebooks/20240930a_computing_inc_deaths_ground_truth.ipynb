```python
df_nyt = df_nyt.sort_values(['Fips','Date'])

dffs = []
for fips in tqdm(df_nyt.Fips.unique()):
    
    dff = df[df.Fips==fips]
    dff.reset_index(inplace=True, drop=True)
    
    inc_deathss = []
    for i in range(len(dff)):
        if i==0:
            inc_deaths = np.nan
        else:            
            inc_deaths = dff.True_cum_deaths.values[i] - dff.True_cum_deaths.values[i-1]
        inc_deathss.append(inc_deaths)
            
    dff['True_inc_deaths'] = inc_deathss
#     break
    
    dffs.append(dff)
```

```python
df = pd.concat(dffs)
df.dropna(inplace=True)
df
```
